### Writing into File

```python
import csv

fields = ['RollNo','Name','Branch']
features=[]

with open ('C:/Users/ml18/Desktop/76/FODS-Sem5-main/data/wroteFile.csv','w',newline="") as outFile:
    fileObj = csv.writer(outFile)
    fileObj.writerow(fields)                             #wrote one row of fields
    
    for i in range (3):
        feature = [int(input()) , input() , input()]
        features.append(feature)                        #took multiple features from user and appended into "features"
    
    fileObj.writerows(features)                          #wrote the fetures list to csv file
```
